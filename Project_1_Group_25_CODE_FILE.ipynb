## This is the code for MLR with PCA

```python
#importing required liabraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#importing dataset
ds=pd.read_csv("Admission_Predict_Ver1.1.csv")
#checking if categorical values or NA values present
print(ds.info())
```

```python
# as there are no NA or categorical values no need for imputing values or encoding(i.e. EDA)
#first 5 values
ds.head()
```

```python
#last 5 values
ds.tail()
```

```python
#extracing x and y values
#as serial number is not important or not a feature which needs for y prediction
#so excluding serial number in x
x=ds.iloc[:,1:-1].values
y=ds.iloc[:,-1].values
print(x)
print(y)
```

```python
#as all data should be on same scale so we do
#Feature scaling on x 
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X = sc.fit_transform(x)
print(X)
```

```python
#explained variance
from sklearn.decomposition import PCA
pca= PCA().fit(X)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumelative explained variance')
plt.grid()
plt.show()
#determining n_components value
pca= PCA(n_components=2)
X_pca=pca.fit_transform(X)
print(pca.explained_variance_ratio_)
X_pca.shape
#as fisrst two features have total variance more than 75% so we take n_components value as 2
```

```python
#visualizing output
plt.figure(figsize=(8,7))
plt.scatter(X_pca[:,0], X_pca[:,1], c=y, cmap='plasma')
plt.xlabel('First Principal Component', fontsize=15)
plt.ylabel('Second Principal Component', fontsize=15)
plt.title('PCA output', fontsize=20)
```

```python
#now doing multiple linear regression
#Splitting x and y into testing and training set
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
x_tr, x_te, y_tr, y_te = train_test_split(x,y,test_size=0.2,random_state=0)
print('X train : \n',x_tr)
mlr= LinearRegression()
mlr.fit(x_tr,y_tr)
```

```python
# predict the output and look for the errors and calculating r2 score
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
y_pred= mlr.predict(x_te)
print('y pred :\n',y_pred)
print('-----------------------------------------------------------------')
print('y test :\n',y_te)
```

```python
mse = mean_squared_error(y_te, y_pred)
mae = mean_absolute_error(y_te, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_te, y_pred)
print('mse =\t',mse)
print('rmse =\t',rmse)
print('mae =\t',mae)
print('r2 =\t',r2)
```

```python
plt.figure(figsize=(5,5))
plt.scatter(x_te[:,1], y_te, c='red', label='actual')
plt.scatter(x_te[:,1], y_pred, c='blue', label='predicted')
plt.legend()
plt.title('Actual vs Predicted', fontsize=20)
plt.xlabel('Input', fontsize=15)
plt.ylabel('Output', fontsize=15)
```

```python
#GRE Score and TOEFL Score are PCA components with highest variance 
#GRE Score with predicted chance of admit
plt.figure(figsize=(6,6))
plt.scatter(x[:,0], y, c='red', label='actual')
plt.scatter(x_te[:,0], y_pred, c='blue', label='predicted')
plt.legend()
plt.title('Actual vs Predicted', fontsize=20)
plt.xlabel('Input', fontsize=15)
plt.ylabel('Output', fontsize=15)
```

```python
#TOEFL Score with predicted chance of admit
plt.figure(figsize=(6,6))
plt.scatter(x[:,1], y, c='red', label='actual')
plt.scatter(x_te[:,1], y_pred, c='blue', label='predicted')
plt.legend()
plt.title('Actual vs Predicted', fontsize=20)
plt.xlabel('Input', fontsize=15)
plt.ylabel('Output', fontsize=15)
```
